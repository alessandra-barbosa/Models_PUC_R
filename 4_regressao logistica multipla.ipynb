```r
class(Dados$Classe)
Dados$Classecat <- as.factor(Dados$Classe)
class(Dados$Classecat)
```

```r
modelo5=glm(Dados$Sobrevivente~Dados$Sexo+Dados$Idade+Dados$Irmãos+Dados$Pais+Dados$Classecat,
family=binomial(link="logit"));modelo5
summary(modelo5)
modelo5$coefficient
```
